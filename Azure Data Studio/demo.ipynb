## üìå Creating your own extension
---
This is a more advanced topic, so we will show you how to create an extension really quick.
if you want to know more about extensions, and follow a more detailed tutorial, please check my articles at:

How to create a Dashboard extension > https://www.mssqltips.com/sqlservertip/6023/creating-dashboard-extensions-for-azure-data-studio/

How to create a Typescript extension > https://www.mssqltips.com/sqlservertip/6047/create-azure-data-studio-extension-with-typescript/

For now, we will take our Widget demo and packet it as an extension.

Prerequisites **(you can check how to install those on any of my articles)**:
 - Node.js
 - VSCode
 - Azure Data Studio debug extension for VSCode
 - Yeoman Extension generator (using npm)
 - VSCE package (using npm), this is used for packaging the .vsix file

Steps:

- Open VSCode
- On a terminal run: <br/><b>yo azuredatastudio</b>
- Select dashboard extension type
- Complete the wizard with the required info
- Navigate to your template by running: 
    <br/><b>cd yourExtensionName<br/>
     code .</b>

- Open query.sql file and add your query (for our example contents of Script3.sql)
- Open package.json file, navigate to üìù **dashboard.insights** and modify the **type** property with the contents of the **type** property on our widget
- Open launch.json file, check if *azuredatastudio* is set on **"runtimeExecutable"**
- Debug it!
- When you are ok with the extension, we will create the .vsix file
- Update readme.md to skip warnings created by vsce
- open package.json and update the vscode engine to a lower number (if no, the vsix import will fail)
- Run the following command on a terminal:
    <br/><b>vsce packate</b>
- If everything is ok, you will be able to import the created .vsix file into Azure Data Studio.
